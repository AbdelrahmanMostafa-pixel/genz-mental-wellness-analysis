```python
# numeric columns
num_cols = df.select_dtypes(include=np.number).columns

for col in num_cols:
    plt.figure()
    sns.histplot(df[col], kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
# categorical columns
cat_cols = df.select_dtypes(include='object').columns

for col in cat_cols:
    plt.figure()
    df[col].value_counts().plot(kind='bar')
    plt.title(f"Count of {col}")
    plt.show()
```

```python
# correlation
plt.figure()
sns.heatmap(df[num_cols].corr(), annot=False)
plt.title("Correlation Matrix")
plt.show()
```

```python
# example relationships
sns.boxplot(x="Gender", y="Social_Comparison_Index", data=df)
plt.show()

sns.scatterplot(
    x="Daily_Sleep_Hours",
    y="Night_Scrolling_Frequency",
    data=df,
    alpha=0.6
)

sns.regplot(
    x="Daily_Sleep_Hours",
    y="Night_Scrolling_Frequency",
    data=df,
    scatter=False
)

plt.title("Sleep vs Night Scrolling")
plt.show()


df[["Daily_Sleep_Hours", "Night_Scrolling_Frequency"]].corr()
```

```python
# sleep vs stress
df.groupby(pd.cut(df["Daily_Sleep_Hours"], bins=5))["Sleep_Quality_Score"].mean()
```

```python
# social media vs anxiety
df.groupby(pd.cut(df["Daily_Social_Media_Hours"], bins=5))["Social_Comparison_Index"].mean()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

# features and target
X = df.select_dtypes(include='number').drop(columns=["Motivation_Level"])
y = (df["Motivation_Level"] > df["Motivation_Level"].mean()).astype(int)

# split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# define model
model = RandomForestClassifier()

# train
model.fit(X_train, y_train)
```

```python
importances = pd.Series(model.feature_importances_, index=X.columns)
importances.sort_values().plot(kind='barh')
plt.title("Feature Importance")
plt.show()
```
